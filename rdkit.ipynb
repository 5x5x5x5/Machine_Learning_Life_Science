````python
from rdkit import Chem
````

````python
m = Chem.MolFromSmiles('Cc1ccccc1')
````

````python
m
````

````python
Chem.Kekulize(m)
````

````python
Chem.MolToSmiles(m,kekuleSmiles=True)
````

````python
from rdkit.Chem import Draw
````

````python
fig = Draw.MolToMPL(m)
````

````python
fig
````
